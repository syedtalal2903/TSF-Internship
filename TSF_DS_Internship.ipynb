```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('http://bit.ly/w-data',index_col=False)
df
```

```python
df.head()
df.shape
```

```python
df.describe()
```

```python
df.info()
df.isnull().sum()
```

```python
df=df.drop_duplicates()
df
```

```python
df.shape
df.describe()
```

```python
X=df.iloc[:,0].values
y=df.iloc[:,1].values
print(X)
print(y)
```

```python
plt.scatter(X,y)
plt.title('Hours vs Scores')
plt.xlabel('Hours Prepared')
plt.ylabel('Score Obtained')
plt.show()
```

```python
X=X.reshape(-1,1)
y=y.reshape(-1,1)
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
pred=model.predict(xtest)
```

```python
plt.scatter(X,y)
plt.xlabel("Hours Studied")
plt.ylabel('Score Obtained')
plt.title("Fitting")
plt.plot(X,model.coef_*X+model.intercept_,color='orange')
plt.show()
```

```python
plt.scatter(xtrain,ytrain,color='orange')
plt.scatter(xtest,ytest,color='green')
plt.scatter(xtest,pred,color='red')
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(ytest,pred,squared=False))
```

```python
prediction=model.predict([[9.30]])
prediction
```
